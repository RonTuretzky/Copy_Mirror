```python
from monday import MondayClient
key_migal= ""
migal_client = MondayClient(key_migal)
```

```python
def fetch_mirror_items(board_id,mirror_col_id,client):
	items = client.boards.fetch_items_by_board_id(board_id)
	id_dictionary_tuples = [(item['id'],item['column_values']) for item in [item for item in items['data']['boards'][0]["items"]]]
	columns = [tup for tup in [(tup[0],tup[1][i]) for tup in id_dictionary_tuples for i in range(len(tup[1]))] if mirror_col_id in tup[1].values()]
	values = [(value[0],value[1]['text']) for value in columns if value[1]['text']!='' ]
	return values
```

```python
def copy_mirror_col_to_other_col(board_id,mirror_col_id,target_col_id,target_col_type,client):
	mirrored_items = fetch_mirror_items(board_id,mirror_col_id,client)
	for item in mirrored_items:
		if target_col_type == 'text':
			out =client.items.change_item_value(board_id,item[0],target_col_id,item[1])
			print(out)
		if target_col_type == 'date':
			out =client.items.change_item_value(board_id,item[0],target_col_id,{"date":item[1],"time":"00:00:00"})
			print(out)
		if target_col_type == 'status':
			out =client.items.change_item_value(board_id,item[0],target_col_id,{"label":item[1]})
			print(out)
```

```python
def copy_mirror_cols_to_other_cols(mirrors,targets,types,client,board_id):
	for mirror_id,target_id,target_type in zip(mirrors,targets,types):
		copy_mirror_col_to_other_col(board_id,mirror_id,target_id,target_type,client)
```
